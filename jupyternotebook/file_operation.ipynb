```python
def find_listdir(file_path):
    ls_0=os.listdir(file_path)
    ls_result_file=[]
    ls_result_filepath=[]
    for i in ls_0:
        if os.path.isdir(os.path.join(file_path,i)):
            ls_result_file.extend(find_listdir(os.path.join(file_path,i))[0])
            ls_result_filepath.extend(find_listdir(os.path.join(file_path,i))[1])
        else:
            split_f=os.path.splitext(i)[1]
            if split_f.startswith('.'):
                format_f=split_f[1:]
                if format_f in ['jpg', 'jpeg', 'png', 'gif', 'bmp','txt', 'csv', 'json', 'xml','pdf','avi', 'mp4', 'mov', 'wmv', 'mkv','doc', 'docx', 'ppt', 'pptx', 'xls', 'xlsx']:
                    ls_result_file.append(i)
                    ls_result_filepath.append(os.path.join(file_path,i))
    return ls_result_file,ls_result_filepath
```

```python
# 根据文件夹名称对里面子文件进行统一改名
def rename_fun1(file_path):
    ls_file=os.listdir(file_path)
    ls_file_new=[]
    for i in ls_file:
        if '.' not in i:
            count=1
            path_i=file_path+'\\'+i
            ls_file_i=os.listdir(path_i)
            base_name=re.findall('\D+',i)[0]

            for a in ls_file_i:
                name_a=base_name+'_'+str(count)+'.jpg'
                count+=1
                os.rename(path_i+'\\'+a,path_i+'\\'+name_a)
    
    for i in ls_file:
        if '.' not in i:
            path_i=file_path+'\\'+i
            ls_file_i=os.listdir(path_i)
            ls_file_new.extend(ls_file_i)
    return ls_file_new
```
