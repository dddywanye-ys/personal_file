```python
# dataframe字段值拼接
def data_join(ls,dataframe):
    ls_result = []
    for i in ls:
        ls_i = dataframe[i].unique().tolist()
        if len(ls_i) >1:
            print('error',i,ls_i)
        val_i = ls_i[0]
        if val_i == 'xx':
            continue
        else:
            ls_result.append(val_i)
    return ls_result
```

```python
r_1 = lambda x,y:x+y
code_ls = list(map(r_1, ['code_']*6,[str(x) for x in range(1,7)]))
name_ls = list(map(r_1, ['name_']*6,[str(x) for x in range(1,7)]))
sk_code_ls = list(map(r_1, ['sk_code_']*6,[str(x) for x in range(1,7)]))
sk_name_ls = list(map(r_1, ['sk_name_']*6,[str(x) for x in range(1,7)]))
ls_code_total = []
ls_total = []
count = 1
for i in range(len(name_ls)):
    print(count)
    # ls_i = []
    ls_code_i = df_total[code_ls[i]].unique().tolist()
    ls_code_i = [x for x in ls_code_i if type(x) != float and x != 'xx']
    # print(len(ls_i))
    for j in ls_code_i:
        ls_j = [j]
        df_j = df_total.loc[df_total[code_ls[i]] == j]
        df_j_ori = df_ori.loc[df_ori['gb_code'] == j]
        
        # 产品名称
        ls_name_j = df_j[name_ls[i]].unique().tolist()
        if len(ls_name_j) > 1:
            print(j,ls_name_j)
        name_j = ls_name_j[0]
        ls_j.append(name_j)
        
        # 产品父级及更高级索引信息
        if i >= 1:
            c = 1
            parent_code_ls = df_j[code_ls[i-c]].unique().tolist()
            parent_code = parent_code_ls[0]
            while parent_code == 'xx':
                c+=1
                parent_code_ls = df_j[code_ls[i-c]].unique().tolist()
                parent_code = parent_code_ls[0]
            parent_code_index = ';'.join(data_join(code_ls[:i],df_j))
        else:
            parent_code_ls = ['']
            parent_code = ''
            parent_code_index = ''
        if len(parent_code_ls) >1:
            print(j,parent_code_ls)
        ls_j.append(parent_code)
        ls_j.append(parent_code_index)
        
        # 产品等级信息
        level_j = i+1
        ls_j.append(level_j)
        
        # 产品所属行业信息
        if len(df_j_ori) == 0:
            ls_j.extend(['',''])
        else:
            ind_code = df_j_ori['industry_code'].unique().tolist()[0]
            ind_name = df_j_ori['ind_name'].unique().tolist()[0]
            if type(ind_name) == float:
                ind_name = ''
            ls_j.append(ind_code)
            ls_j.append(ind_name)
        
        # 产品所对应的数库产品名称
        sk_code_ls_j = df_j[sk_code_ls[i]].unique().tolist()
        sk_name_ls_j = df_j[sk_name_ls[i]].unique().tolist()
        if len(sk_code_ls_j) >1:
            sk_code_str_j = ';'.join(sk_code_ls_j)
            sk_name_str_j = ';'.join(sk_name_ls_j)
        else:
            sk_code_str_j = sk_code_ls_j[0]
            sk_name_str_j = sk_name_ls_j[0]
        ls_j.append(sk_code_str_j)
        ls_j.append(sk_name_str_j)
        
        # 产品所提取的关键匹配信息、更新状态、更新时间
        if len(df_j_ori) == 0:
            ls_j.extend(['','','数库产品融入'])
            ls_total.append(ls_j)
        else:
            label_item = df_j_ori['Label_item'].tolist()
            label_item_final = df_j_ori['Label_item_jiang'].tolist()
            for a in range(len(label_item_final)):
                ls_j_a = ls_j.copy()
                if type(label_item[a]) == float:
                    label_item[a] = ''
                ls_j_a.append(label_item[a])
                ls_j_a.append(label_item_final[a])
                if len(j) > 10:
                    ls_j_a.append('数库产品融入')
                else:
                    ls_j_a.append('国标产品')
                ls_total.append(ls_j_a)
    # ls_code_total.extend(ls_i)
    count+=1
```
