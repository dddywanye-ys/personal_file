```python
import time
import arcpy
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import csv
import sys
pd.set_option("display.max_columns", 100)
```

```python
df=pd.read_csv('E:\\project_data\\25issue\\'+'boundary_location_.csv')
df_poi=pd.DataFrame()
ls_uid=df['UID'].unique().tolist()
for i in ls_uid:
    df_i=df.loc[df['UID']==i].reset_index(drop=True)
    df_xy=pd.DataFrame()
    if len(df_i)>1 or len(df_i.loc[df_i['polyline'].notnull()])==0:
        print(df_i)
        continue
    for a in range(len(df_i)):
        points=df_i.loc[a,'polyline']
        ls_points=[[float(y) for y in x.split(',')] for x in points.split(';')]
        df_points=pd.DataFrame(ls_points,columns=['x','y'])
        df_xy=pd.concat([df_xy,df_points],axis=0)
    df_i=df_i.drop(labels=['polyline'],axis=1)
    df_xy.loc[:,'UID']=i
    df_i_=pd.merge(left=df_i,right=df_xy,on=['UID'],how='outer')
    df_poi=pd.concat([df_poi,df_i_],axis=0)
```

```python
# -*- coding: utf-8 -*-
import arcpy
import io
import ZB_transformimport os
from arcpy import da
from arcpy import env
from imp import reload
import csv
# import sys# reload(sys)# sys.setdefaultencoding('utf8')  # 设置默认编码格式为'utf-8'
env.overwriteOutput = True# 基于公交信息文本文件生成公交shpprint "qiguai"
outputpath = "D:\\output\\gongjiao\\shape\\"                     #输出文件夹
shp_output = outputpath + "gongjiao_cs2.shp"                      #输出的文件名称
busInformationCSV = "D:\\output\\gongjiao\\data\\line.csv"       #读取文件位置# 投影参照文件
wgs84file = outputpath + "wgs84.prj"                            #wgs84地理坐标文件
wgs84projectfile = outputpath + "wgs84project.prj"              #wgs84投影坐标文件
wgs84 = arcpy.SpatialReference(wgs84file)                       #设置坐标参照
wgs84project = arcpy.SpatialReference(wgs84projectfile)         #投影参照print "创建shp"# 生成公交shp
arcpy.CreateFeatureclass_management(os.path.dirname(shp_output),os.path.basename(shp_output),'POLYLINE',"","","",wgs84project)       #创建一个空白的线要素类# 给shp添加字段属性
arcpy.AddField_management(shp_output,'Fangxiang','Text')
arcpy.AddField_management(shp_output,'Bianhao','Text')
arcpy.AddField_management(shp_output,"BusType",'Text')
arcpy.AddField_management(shp_output,'Name','Text')
arcpy.AddField_management(shp_output,"S_stop",'Text')
arcpy.AddField_management(shp_output,"E_stop",'Text')
arcpy.AddField_management(shp_output,"StartTime",'Text')
arcpy.AddField_management(shp_output,"EndTime",'Text')
arcpy.AddField_management(shp_output,"BusCompany",'Text')
arcpy.AddField_management(shp_output,"Distance",'Text')
arcpy.AddField_management(shp_output,"basic_p",'Text')
arcpy.AddField_management(shp_output,"Total_p",'Text')
arcpy.AddField_management(shp_output,"StationNum",'Text')
arcpy.AddField_management(shp_output,"Length",'Double')
fields =['SHAPE@','Fangxiang','Bianhao','BusType',"Name",'S_stop','E_stop','StartTime','EndTime','BusCompany','Distance','basic_p','Total_p','StationNum','Length']
cur = da.InsertCursor(shp_output,fields)         #插入行
array = arcpy.Array()
# 读取公交信息文件print "读取信息"
csv_information = csv.reader(open(busInformationCSV,'rt'))
buslines = []
for csv in csv_information:
    buslines.append(csv)
    for j in range(1,len(buslines)):
        base_data = buslines[j]    
        print("-------这里是基础信息-----------------")    # print base_data
        Fangxiang = base_data[0]    
        Bianhao = base_data[1]    
        BusType = base_data[2]    
        Name = base_data[3]    
        S_stop = base_data[4]    
        E_stop = base_data[5]    
        startTime = base_data[6]    
        endTime = base_data[7]    
        busCompany = base_data[8]    
        Distance = base_data[9]    
        basic_p = base_data[10]    
        Total_p = base_data[11]    
        stationNum = base_data[12]    
        print startTime    
        print "--------------这里是坐标点的信息------------------------"    
        points_data = base_data[13]    
        points_data = points_data.split(";")    
        for i in range(1,len(points_data)):    #将polyline的坐标点循环遍历取出来        
            print i        
            print points_data[i]        
            Lon = float(points_data[i].split(",")[0])        
            Lat = float(points_data[i].split(",")[1])        
            oldLon,oldLat = ZB_transform.gcj02_to_wgs84(Lon,Lat)   #将坐标转成WGS84        
            print oldLon,oldLat# # break        
            pnt = arcpy.Point(oldLon,oldLat)      #对每一个坐标点生成一个point        
            array.add(pnt)          #将每一个point生成一个一维的序列    
            polyline = arcpy.Polyline(array,wgs84)     #将点集生成线，并定义坐标系
            polylinePeoject = polyline.projectAs(wgs84project)   #投影    
            array.removeAll()   #生成一个线要素后需要清空，以便后续的数据写入    #这里是游标插入的行数据，分别是刚刚生成的线几何要素，以及之前的公交基础数据，生成一个列表    
            newFields = [polylinePeoject,Fangxiang,Bianhao,BusType,Name,S_stop,E_stop,startTime,endTime,busCompany,Distance,basic_p,Total_p,stationNum,polylinePeoject.length]
            cur.insertRow(newFields)    #一行一行的写入数据
        del cur
```
